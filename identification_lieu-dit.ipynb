```python
path_topo = "./data/bdd_topo_france.json"
out_path = "./data/bdd_topo_991001184.jsonl"
prefix = "991001184"

# 1) Filtre principal par préfixe
with open(path_topo, "rb") as f, open(out_path, "w", encoding="utf-8") as out:
    for obj in ijson.items(f, "item"):
        code = obj.get("code_topo")
        if code and code.startswith(prefix):
            out.write(json.dumps(obj, ensure_ascii=False) + "\n")
```

```python
from ijson.common import IncompleteJSONError

path_topo = "./data/bdd_topo_france.json"
out_path = "./data/bdd_topo_991009384.jsonl"
prefix = "991009384"

with open(path_topo, "rb") as f, open(out_path, "w", encoding="utf-8") as out:
    try:
        for obj in ijson.items(f, "item"):
            code = obj.get("code_topo")
            if code and code.startswith(prefix):
                out.write(json.dumps(obj, ensure_ascii=False) + "\n")
    except IncompleteJSONError:
        pass  # fin de fichier cassée, on garde ce qui a été lu
```

```python
import ijson
from ijson.common import IncompleteJSONError

path_topo = "./data/bdd_topo_france.json"
prefix = "991009384"

total = 0
matched = 0

with open(path_topo, "rb") as f:
    try:
        for obj in ijson.items(f, "item"):
            total += 1
            code = obj.get("code_topo")
            if code and code.startswith(prefix):
                matched += 1
    except IncompleteJSONError:
        pass

print("total lus:", total)
print("matches:", matched)
```
