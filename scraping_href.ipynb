```python
import re
from pathlib import Path

text = Path("./data/vaucluse_page_principale.html").read_text(encoding="utf-8", errors="ignore")

# <a ... href="/archives/show/FRAD..._de-123"> ... Possde des mdias associs ... </a>
# Exclut tt1/tt2/... en imposant _de-<num> avant la fin du href
pattern = re.compile(
    r'<a\b[^>]*href="(?P<href>/archives/show/[^"]*_de-\d+)"[^>]*>'
    r'(?:(?!</a>).)*?Possde\s+des\s+mdias\s+associs(?:(?!</a>).)*?</a>',
    re.IGNORECASE | re.DOTALL
)

hrefs = [m.group("href") for m in pattern.finditer(text)]

# ddoublonnage en conservant l'ordre
seen = set()
hrefs = [h for h in hrefs if not (h in seen or seen.add(h))]

print("\n".join(hrefs))
```

```python
html = Path("./data/html_FRAD084_07FI002_0013.html").read_text(encoding="utf-8", errors="ignore")

def extract_one(pattern: str) -> str:
    m = re.search(pattern, html, flags=re.IGNORECASE | re.DOTALL)
    return re.sub(r"<[^>]+>", "", m.group(1)).strip() if m else ""

def extract_all(pattern: str):
    return re.findall(pattern, html, flags=re.IGNORECASE | re.DOTALL)

# 1) dernier href "viewer" qui contient le nom du .jpg (lien complet)
viewer_links = extract_all(r'href="(https?://[^"]*?/FRAD084_[^"/]+\.jpg[^"]*)"')
lien_photo = viewer_links[-1] if viewer_links else ""

# 2) Mtadonnes demandes (regex bases sur ton HTML)
dates = extract_one(r'<section[^>]*property="dc:date"[^>]*>.*?<div>(.*?)</div>\s*</section>')

dimensions = extract_one(r'<h4>\s*Dimensions\s*</h4>\s*(.*?)</span>')
donnees_tech = extract_one(r'<h4>\s*Donnes techniques\s*</h4>\s*(.*?)</span>')

infos_entree = extract_one(r'<section[^>]*class="acqinfo"[^>]*>.*?<p>(.*?)</p>\s*</section>')

# Matire / difice : pas visibles dans lextrait fourni -> renvoie "" si absent
matiere = extract_one(r'<strong>\s*Mati[e]re\s*:</strong>\s*<a[^>]*>(.*?)</a>')
edifice = extract_one(r'<strong>\s*[E]difice\s*:</strong>\s*<a[^>]*>(.*?)</a>')

nom_geo = extract_one(r'<strong>\s*Nom g[e]ographique\s*:</strong>.*?<a[^>]*>(.*?)</a>')

result = {
    "Lien_photo": lien_photo,
    "Dates": dates,
    "Description physique": "",  # dans ton extrait, il ny a pas de texte ddi hors Dimensions / Donnes techniques
    "Dimensions": dimensions,
    "Donnes techniques": donnees_tech,
    "Informations sur les modalits d'entre": infos_entree,
    "Matire": matiere,
    "Nom gographique": nom_geo,
    "difice": edifice,
}

print(result)
```
